# January temperature from US cities

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

## load temperature data

```python
atl=np.genfromtxt('atlanta_temperature.tsv',delimiter='\t')
bos=np.genfromtxt('bos_temp.tsv',delimiter='\t')
wil=np.genfromtxt('082320WilmingtonOHDataHW1.tsv',delimiter='\t')
hon=np.genfromtxt('hontemp.tsv',delimiter='\t')
chi=np.genfromtxt('Chicago Past Temp.tsv',delimiter='\t')
bld=np.genfromtxt('BoulderCO_temp.tsv',delimiter='\t')
```

```python
year=atl[:,0]
jan=atl[:,1]
ind=np.where(year>=1950)
A=np.zeros((70,4))
A[:,0]=jan[ind]-np.nanmean(jan[ind])
```
