# January temperature from US cities

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('dark_background')
```

## load temperature data

```python
atl=np.genfromtxt('DATA/atlanta_temperature.tsv',delimiter='\t')
bos=np.genfromtxt('DATA/bos_temp.tsv',delimiter='\t')
wil=np.genfromtxt('DATA/082320WilmingtonOHDataHW1.tsv',delimiter='\t')
hon=np.genfromtxt('DATA/hontemp.tsv',delimiter='\t')
```

## Atlanta : 1950-

```python
year=atl[:,0]
jan=atl[:,1]
ind=np.where(year>=1950)
A=np.zeros((4,70))
A[0,:]=jan[ind]-np.nanmean(jan[ind])
```

## Boston

```python
year=bos[:,0]
jan=bos[:,1]
ind=np.where(year>=1950)
A[1,:]=jan[ind]-np.nanmean(jan[ind])
```

## Wilmington OH

```python
year=wil[:,0]
jan=wil[:,1]
ind=np.where(year>=1950)
A[2,:]=jan[ind]-np.nanmean(jan[ind])
```

## Honolulu HI

```python
year=hon[:,0]
jan=hon[:,1]
ind=np.where(year>=1950)
A[3,:]=jan[ind]-np.nanmean(jan[ind])
```

## Quick look at the data

```python
plt.plot(year[ind],np.transpose(A));
plt.legend(("ATL","BOS","WIL","HON"));
plt.xlabel('year')
plt.ylabel('tempearture, degF')
```

```python
plt.plot(A[0,:],A[1,:],'o');
plt.xlabel('Atlanta temp')
plt.ylabel('Boston temp');
```

```python
plt.plot(A[0,:],A[2,:],'x');
plt.xlabel('Atlanta temp')
plt.ylabel('Wilmington temp');
```

```python
plt.plot(A[0,:],A[3,:],'v');
plt.xlabel('Atlanta temp')
plt.ylabel('Honolulu temp');
```

## Calculate the covariance matrix : Cx

```python
C=np.cov(A)
```

```python
df=pd.DataFrame(C,index=['ATL','BOS','WIL','HON'],columns=['ATL','BOS','WIL','HON'])
df.style
```

## Diag(Cx) is the variance

```python
s=np.sqrt(np.diag(C))
print(s)
```

## correlation matrix : R

```python
R=np.corrcoef(A)
df=pd.DataFrame(R,index=['ATL','BOS','WIL','HON'],columns=['ATL','BOS','WIL','HON'])
df.style
```
