# Week 3 exercise - hypothesis testing - 

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

## 1. State the confidence level

```python
CL=0.95
```

## 2. State the null hypothesis and alternative hypothesis
H0: Mean temp from last 10 years in Boston is NOT significantly warmer than long-term mean

H1: Mean temp from last 10 years in Boston is significantly warmer than long-term mean

```python
data=np.genfromtxt('bos_temp.tsv',delimiter='\t')
temp=data[:,7]
temp[temp==-999]=np.nan
year=data[:,0]
plt.plot(year,temp)
plt.xlabel('year')
plt.ylabel('boston jan temp, deg F');
```

```python
mu=np.nanmean(temp);
M=len(temp)
m10=np.nanmean(temp[M-10:M+1])
s10=np.nanstd(temp[M-10:M+1],ddof=1)
print('long term mean is '+str(round(mu,2))+' deg F')
print('the last 10 year mean is '+str(round(m10,2))+' deg F')
```

## 3. State the statistics used

We use Student's t-distribution with one-tail test

```python
N=10
df=N-1
t=np.arange(-5,5,0.1)
tdist=stats.t.pdf(t,df)
plt.plot(t,tdist)
plt.xlabel('t')
plt.ylabel('t-distribution');
```

## 4. Calculate the critical region

```python
tcrit=stats.t.ppf(CL,df)
print('critical t value is ',str(round(tcrit,2)))
```

## 5. evaluate if the data is within the critical region

```python
tval=(m10-mu)/(s10/np.sqrt(N-1))
print('t value is ',str(round(tval,2)))
```
